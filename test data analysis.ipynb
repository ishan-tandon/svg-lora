```python
# In a new cell, analyze your training data
import pandas as pd
import re

df = pd.read_csv("csv/train.csv")

def has_repetition(svg, window=40, threshold=3):
    for i in range(len(svg) - window):
        snippet = svg[i:i+window]
        if svg.count(snippet) >= threshold:
            return True
    return False

# Check how many training samples have loops
df['has_loop'] = df['svg'].apply(lambda x: has_repetition(str(x)))
print(f"Total samples: {len(df)}")
print(f"Samples with loops: {df['has_loop'].sum()}")
print(f"Clean samples: {(~df['has_loop']).sum()}")
```
